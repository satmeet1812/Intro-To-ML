```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MR data - Compiled Data Set.csv')
df
```

```python
df.drop(['Title', 'Text'], axis=1, inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
df.columns = (i for i in range(0, len(df.columns)) )
df
```

```python
df.describe()
```

```python
dic = {0:0,
       'Property Disputes':1,
       'Family Dispute':2,
       'Petty Quarrels':3,
       'Money Disputes':4, 
       'Personal Vendetta':5, 
       'Love Affairs':6, 
       'Casteism':7,
       'Unknown reasons':8}
```

```python
df[1] = df[1].apply(lambda x : dic[x] )
```

```python
df.head(20)
```

```python
DF = df[[8, 9, 10]].copy(deep=True)
DF
```

```python
df = df.rename(columns={ 9:11, 10:12 })
```

```python
df.insert(9, 9, 0)
df.insert(10, 10, 0)
```

```python
df.head(20)
```

```python
for i in range(0, len(df)):
    if ',' in str(df.iloc[i,8]):
        spl = df.iloc[i,8].split(',')
        df.iloc[i,8] = int(spl[0])
        df.iloc[i,9] = int(spl[1])
        if len(spl)>2:
            df.iloc[i,10] = int(spl[2])
```

```python
Ghaziabad = df.head(167)
Ghaziabad
```

```python
Kanpur = df.iloc[167:264]
Kanpur = Kanpur.reset_index()
Kanpur = Kanpur.drop(['index'],axis=1)
Kanpur
```

```python
Lucknow = df.iloc[264:len(df)]
Lucknow = Lucknow.reset_index()
Lucknow = Lucknow.drop(['index'],axis=1)
Lucknow
```

```python
cases = cases_G = cases_K = cases_L = [ 0 for i in range(0,8)]
cases
```

```python
for i in range(0,len(df)):
    if df.iloc[i,1] != 0:
        cases[df.iloc[i,1] - 1] += 1
cases
```

```python
for i in range(0,len(Ghaziabad)):
    if Ghaziabad.iloc[i,1] != 0:
        cases_G[Ghaziabad.iloc[i,1] - 1] += 1
cases_G
```

```python
for i in range(0,len(Kanpur)):
    if Kanpur.iloc[i,1] != 0:
        cases_K[Kanpur.iloc[i,1] - 1] += 1
cases_K
```

```python
for i in range(0,len(Lucknow)):
    if Lucknow.iloc[i,1] != 0:
        cases_L[Lucknow.iloc[i,1] - 1] += 1
cases_L
```

```python
Murder_reasons = ['Property Disputes','Family Dispute','Petty Quarrels','Money Disputes', 'Personal Vendetta', 'Love Affairs', 'Casteism','Unknown reasons']
Murder_reasons
```

```python
plt.figure(figsize=(20,10))
plt.bar(Murder_reasons, cases, 0.5, color='cyan')
plt.title('Murder Reasons', fontsize=20)
plt.xlabel("Reasons", fontsize=20)
plt.ylabel("Cases", fontsize=20)
```

```python
plt.pie(cases, labels=Murder_reasons, autopct='%.0f%%')
plt.title('Murder Reasons', fontsize=20)
```

```python
plt.figure(figsize=(20,21))

plt.subplot(3,1,1)
plt.plot(Murder_reasons, cases_G, 'm*')
plt.title('Murder Reasons in Ghaziabad', fontsize=25)
plt.xlabel("Reasons", fontsize=20)
plt.ylabel("Cases", fontsize=20)

plt.subplot(3,1,2)
plt.plot(Murder_reasons, cases_G, 'k^')
plt.title('Murder Reasons in Kanpur', fontsize=25)
plt.xlabel("Reasons", fontsize=20)
plt.ylabel("Cases", fontsize=20)

plt.subplot(3,1,3)
plt.plot(Murder_reasons, cases_G,)
plt.title('Murder Reasons in Lucknow', fontsize=25)
plt.xlabel("Reasons", fontsize=20)
plt.ylabel("Cases", fontsize=20)

plt.suptitle("Murder Reasons in individual cities", fontsize=30)
```

```python
cities = ["Ghaziabad", "Kanpur", "Lucknow"]
plt.pie([sum(cases_G), sum(cases_K), sum(cases_L)], labels=cities)
plt.legend(title="Cities")
```
